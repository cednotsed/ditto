```r
require(ggtree)
require(ggtreeExtra)
require(tidyverse)
require(data.table)
require(ape)
require(ggnewscale)
```

```r
tree <- read.tree("data/trees/mink_n789_201121.audacity_only.v8_masked.tree")

animal_meta <- fread("data/metadata/combined_metadata_201121.audacity_only.tsv") %>%
  select(accession_id, host, clade, lineage, location) %>%
    separate(location, into = c("loc1", "loc2"), sep = " / ") %>%
    mutate(location = paste0(loc1, " / ", loc2)) %>%
  rename(Accession = accession_id) %>%
    as_tibble()
```

```r
animal_meta <- animal_meta[match(tree$tip.label, animal_meta$Accession), ]
animal_meta[is.na(animal_meta$Accession), ] <- "root"
```

```r
dd <- data.frame(id = tree$tip.label,
                 lineage = animal_meta$lineage,
                 location = animal_meta$location,
                 clade = animal_meta$clade)
```

```r
# Plot tree
p <- ggtree(tree)
p <- p %<+% dd
p + new_scale_fill() +
    geom_fruit(geom = geom_tile, aes(fill = lineage), width=0.0015, offset=-0.5, alpha = 0.1) +
    geom_tippoint(aes(color = location)) +
    labs(color = "Location", fill = "Lineage") +
theme_tree2()
```
